

### В поле "rating_grade" записана оценка отзыва по шкале от 1 до 5. Используйте меру $tf-idf$, для того, чтобы найти ключевые слова и биграмы для положительных отзывов (с оценкой 5) и отрицательных отзывов (с оценкой 1)



```python
import json

import bz2
import regex
from tqdm import tqdm
from scipy import sparse
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import pandas as pd
import numpy as np
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%pylab inline
```

```python
from nltk.corpus import stopwords
from string import punctuation
noise = stopwords.words('russian') + list(punctuation)
```

```python
responses = []
with bz2.BZ2File('banki_responses.json.bz2', 'r') as thefile:
    for row in tqdm(thefile):
        resp = json.loads(row)
        if not resp['rating_not_checked'] and (len(resp['text'].split()) > 0):
            responses.append(resp)
```

```python
responses[6]
```

```python
text = [] 
for row in responses:
    text.append(row['text'])
len(text)
```

```python
text_good = []
for row in responses:
    if row["rating_grade"] == 5:
        text_good.append(row['text'])
```

```python
text_good[0]
```

```python
# приведем все слова к нижнему регистру
texts_good_lower_case = [text.lower() for text in text_good]
```

```python
# уберем все служебные символы
only_words_text_good = [re.sub('[\W_]+',' ', text) for text in texts_good_lower_case]
```

```python
# разделим все на слова
texts_good_splitted = [text.split() for text in only_words_text_good]
```

```python
#стопслова уберем

from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('russian'))
without_sw_good = [[word for word in text if word not in stopwords_set] for text in texts_good_splitted]
```

```python
#лемматизация
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
lemmatized_text_good = [[wordnet_lemmatizer.lemmatize(word) for word in text] for text in without_sw_good]
```

```python
all_text_good = []
for text in lemmatized_text_good:
    for word in text:
        all_text_good.append(word)
```

```python
len(text_good)
```

```python
from gensim import corpora

# создадим словарь, в котором присвоем каждому словую свой уникальный id
dictionary_good = corpora.Dictionary(lemmatized_text_good)
dict(dictionary_good)
```

```python
#Создадим на основе словаря модель bag of words, которая присваивает каждому слову (id) количество его вхождений в конкретном документе
#bows_good = [dictionary_good.doc2bow(text) for text in lemmatized_text_good]
bows_good = dictionary_good.doc2bow(all_text_good)
bows_good
```

```python
bow_good = pd.DataFrame(bows_good)
bow_good.columns = ['index', 'occurrences']
bow_good
```

```python
# смэтчим индексы с реальными словами и получим самые частотные слова для хороших отзывов
bow_good['word'] = [dictionary_good[index] for index in bow_good['index']]
sort_bow_good = bow_good.sort_values(by='occurrences', ascending=False)
sort_bow_good.head(10)
```

```python
#теперь сделаем то же самое с помощью  TfidfVectorizer

vec_good_1 = TfidfVectorizer(ngram_range=(1, 1), stop_words = noise)
bow_good_1 = vec_good_1.fit_transform(text_good) # bow -- bag of words (мешок слов)
```

```python
#list(vec_good_1.vocabulary_.items())[:20]
```

```python
#получаем ключевые слова для хороших отзывов
c = bow_good_1.tocoo()
df = pd.DataFrame({'index': c.row, 'word_index': c.col, 'edge_weight': c.data})
sort_df = df.sort_values(by='edge_weight', ascending=False)
df2 =sort_df.head(10)  
df2

```

```python
for index in df2['word_index']:
    print(index)
```

```python
#получаем самые частотные слова для хороших отзывов
for k, v in vec_good_1.vocabulary_.items():
    for index in df2['word_index']:
        if index == v:
            print(k)
    
```

```python
#тоже самое для биграмм
vec_good_bi = TfidfVectorizer(ngram_range=(2, 2), stop_words = noise)
bow_good_bi = vec_good_bi.fit_transform(text_good) # bow -- bag of words (мешок слов)
```

```python
list(vec_good_bi.vocabulary_.items())[:20]
```

```python
#получаем ключевые биграммы для хороших отзывов
c2 = bow_good_bi.tocoo()
df3 = pd.DataFrame({'index': c2.row, 'word_index': c2.col, 'edge_weight': c2.data})
sort_df3 = df3.sort_values(by='edge_weight', ascending=False)
df4 =sort_df3.head(10)  
df4
```

```python
for index in df4['word_index']:
    print(index)
```

```python
#получаем ключевые биграммы для хороших отзывов
for k, v in vec_good_bi.vocabulary_.items():
    for index in df4['word_index']:
        if index == v:
            print(k)
```
